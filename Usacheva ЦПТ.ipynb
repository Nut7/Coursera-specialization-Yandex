```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as  sts
import math
%matplotlib inline
```

Выберите ваше любимое непрерывное распределение (чем меньше оно будет похоже на нормальное, тем интереснее; попробуйте выбрать какое-нибудь распределение из тех, что мы не обсуждали в курсе). 

Возьмем распределение Лапласа

```python
laplac= sts.laplace(5)  #betta=5, alpha=1
```

1. Сгенерируйте из него выборку объёма 1000,
2. постройте гистограмму выборки и нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины (чтобы величины были в одном масштабе, не забудьте выставить у гистограммы значение параметра normed=True).

```python
sample = laplac.rvs(1000)  # так, генерируем выборку объема 1000

x = np.linspace(0,30,100)  #задаем иксу интервал
pdf = laplac.pdf(x)

plt.hist(sample, normed=True,  label='sample histogram') #строим гистограммку

plt.plot(x, pdf, label='theoretical pdf') #строим pdf(теоретическую потность распр-я)

plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

Ваша задача:            
1.  оценить распределение выборочного среднего вашей случайной величины при разных объёмах выборок.     
Для этого при трёх и более значениях n (например, 5, 10, 50) сгенерируйте 1000 выборок объёма n и постройте гистограммы распределений их выборочных средних.

2.  Используя информацию о среднем и дисперсии исходного распределения (её можно без труда найти в википедии), посчитайте значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних. 

Так. Среднее и дисперсия у распределения Лапласа такие:             
     $μ$ = $β$ = 5      
     $σ$ = $2/α^2 $=2                             
Согласно ЦПТ, выборочные средние распределятся нормально со средним x, совпадающим со средним ГС $μ$ и со стандартным отклонением Se:           
                           
$$x = μ = 5 $$           
$$Se= {{σ}/\sqrt(n)} = 2/\sqrt{n}$$           

Поверх каждой гистограммы нарисуйте плотность соответствующего нормального распределения  N(5, 2/sqrt(n))     
(будьте внимательны с параметрами функции, она принимает на вход не дисперсию, а стандартное отклонение).

# n=5

```python
n=5
sample_means = [] #сюда запишем 1000 средних наших выборок
for rvn in range(1000):
    sample_means.append(laplac.rvs(n).mean())  #rvs это походу случайные величины
#и рисуем:
plt.hist(sample_means, 20, normed=True, range=[0, 12], label='sample means histogram')
    
# и вычисляем pdf нормального распределения с параметрами, которые мы рассчитали
norm_rv = sts.norm(5, 2/math.sqrt(n))
x = np.linspace(0, 10, 100)
pdf = norm_rv.pdf(x)
    # и вырисовывается на графике
plt.plot(x, pdf, label='norm theoretical pdf')
    # легенда графика
plt.legend()
plt.title("Sample size %d" % n)
plt.ylabel('fraction of samples $f(x)$')
plt.xlabel('sample mean $x$')
plt.show()
```

# n=10

```python
#все то же самое
n=10
sample_means = []
for rvn in range(1000):
    sample_means.append(laplac.rvs(n).mean())

plt.hist(sample_means, 20, normed=True, range=[0, 12], label='sample means histogram')
    

norm_rv = sts.norm(5, 2/math.sqrt(n))
x = np.linspace(0, 10, 100)
pdf = norm_rv.pdf(x)

plt.plot(x, pdf, label='norm theoretical pdf')

plt.legend()
plt.title("Sample size %d" % n)
plt.ylabel('fraction of samples $f(x)$')
plt.xlabel('sample mean $x$')
plt.show()
```
